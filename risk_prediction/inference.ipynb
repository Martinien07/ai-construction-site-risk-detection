```python
# ==========================================================
# IMPORTS & PATH
# ==========================================================
import sys
sys.path.insert(0, r"D:\ETUDEµ\CITE\ETAPE4\PROJET DE FIN DE SESSION\TOURMANT 1")


from risk_prediction.hse_rule_engin import HSERuleEngine

import json

# Chargement des règles
with open("../rules_identification/config/hse_rules.json", "r", encoding="utf-8") as f:
    rules_config = json.load(f)

# Initialisation du moteur HSE
hse_engine = HSERuleEngine(rules_config)

# Exemple de caractéristiques extraites (fenêtre temporelle)
features = {
    "avg_epi_compliance": 0.72,
    "helmet_compliance_ratio": 0.6,
    "vest_compliance_ratio": 0.95,
    "mask_compliance_ratio": 1.0,
    "min_person_machine_distance": 1.2,
    "time_near_machine": 4.3,
    "crowding_score": 0.4
}

# Activité prédite par le modèle
activity_id = 5  # Travail_près_machine

# Évaluation HSE
result = hse_engine.evaluate(features, activity_id)

# Affichage des résultats
print("Score de risque :", result["risk_score"])
print("Niveau de risque :", result["risk_level"])
print("EPI manquants :", result["missing_epi"])

print("\nRègles violées :")
for rule in result["violated_rules"]:
    print(f"- {rule['rule_id']} | {rule['name']} | {rule['severity']}")
    print(f"  → {rule['recommendation']}")
    if rule["missing_epi"]:
        print(f"  → EPI : {', '.join(rule['missing_epi'])}")




```
